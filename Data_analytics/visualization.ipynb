```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y=[21,345,56,12,5,47,1]
plt.plot(y)
plt.show()
```

```python
plt.figure(figsize=(15,4))
plt.pie(y)
plt.show
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,figsize=(15,5))
plt.show()
```

```python
t=np.arange(-3* np.pi,3* np.pi,0.2)
x=np.sin(t)
y=np.cos(t)
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,2))
ax[0,][0].plot(t,x,'r')
ax[0][1].bar(t,y)
plt.show
```

```python
x=[100,200,300,400,500,600]
y=[23,45,76,89,12,67]
plt.plot(x,y,linewidth=2,color='green',marker='o',markersize=15,markerfacecolor='white',markeredgecolor='red',linestyle='--')
plt.show()
```

```python
x=[100,200,300,400,500,600]
y=[23,45,76,89,12,67]
plt.plot(x,y,'go--',linewidth=2,markersize=15,markerfacecolor='white',markeredgecolor='red')
```

```python
plt.plot(x,y,linewidth=2,color='green',marker='o',markersize=15,markerfacecolor='red',markeredgecolor='white',linestyle='--')
plt.title('Some graph showing something',fontdict={'fontsize':15, 'fontweight':'light','color':'green'})
plt.xlabel('X axis ki value')
plt.ylabel('Y axis ki value', fontdict={'color':'red'})
plt.show()
```

```python
plt.plot(x,y,'go--',linewidth=2,markersize=15,markerfacecolor='white',markeredgecolor='red')
plt.title('Some graph showing something',fontdict={'fontsize':15, 'fontweight':'light','color':'green'})
plt.xlabel('X axis ki value')
plt.ylabel('Y axis ki value', fontdict={'color':'red'})
plt.text(500,80,'This is serious point')
plt.annotate('Peak value',xy=(400,90),xytext=(500,70),arrowprops=dict(facecolor='black',width=1))
plt.show()
```
